```python
#Import the library
import matplotlib.pyplot as plt
%matplotlib inline

class Circle(object):
    def __init__(self,radius=3,color='blue'):
        self.radius = radius
        self.color = color
    #Method
    def add_radius(self,r):
        self.radius = self.radius + r
        return(self.radius)
    #Method

    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0,0),radius = self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()

         
```

```python
#Create an object RedCircle
RedCircle = Circle(10,'red')

#Find out the methods can be used on the object RedCircle
dir(RedCircle)
```

```python
#Print the object attribute radius

RedCircle.radius
```

```python
#Print the object attribute color
RedCircle.color
```

```python
#Set the object attribute radius

RedCircle.radius = 1
RedCircle.radius 
```

```python
#Call the method drawCircle
RedCircle.add_radius(-7)
RedCircle.drawCircle()
```

```python
# Create a new Rectangle class for creating a rectangle object

class Rectangle(object):

    #Constructor
    def __init__(self,width=2,height=3,color='r'):
        self.height = height
        self.width = width
        self.color = color

    #Method
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0,0),self.width,self.height,fc=self.color))
        plt.axis('scaled')
        plt.show()        
```

```python
#Create a new object rectangle

SkinnyBlueRectangle = Rectangle(2,3,'blue')

#Print the object attribute height
SkinnyBlueRectangle.height
```

```python
#Print the object attribute width
SkinnyBlueRectangle.width
```

```python
#Print the object attribute color
SkinnyBlueRectangle.color
```

```python
#Use the drawRectangle method to draw the shape
SkinnyBlueRectangle.drawRectangle()
```

```python
#Exercise of Car dealership's inventory managment system

#TASK 1: Creating python program to represent vehicles using a class. Each car should have attributes for maximum speed and mileage.

class Vehicle:
    color = 'White'
    def __init__(self,max_speed,mileage):
        self.max_speed = max_speed
        self.mileage = mileage
        self.seating_capacity = None
#TASK 2: Update the class with default color for all vehicles, "white".
#TASK 3: Creating methods assigning seating capacity to a vehicle.

    def assign_seating_capacity(self,seat_capacity):
        self.seating_capacity = seat_capacity

#TASK 4: Creating a method to display all the properties of an object of the class
    def specs(self):
        print('Properties of the vehicles: ')
        print(f'Color: {self.color}')
        print(f'Maximum speed: {self.max_speed}')
        print(f'Mileage: {self.mileage}')
        print(f'Seating capacity: {self.seating_capacity}')

```

```python
# TASK 5: 
# Creating 2 objects of Vehicle class object that should have a max speed of 200 kph and mileage of 50000kmpl with 5 seats.
# Max speed of 180 kph and 75000kmpl with 4 seats.
```

```python
first_car = Vehicle(200,50000)
first_car.assign_seating_capacity(5)
first_car.specs()
```

```python
second_car = Vehicle(180,75000)
second_car.assign_seating_capacity(4)
second_car.specs()
```
