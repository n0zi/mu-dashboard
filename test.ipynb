```python
L = ['Michael Jackson',10.1,1982]
L
```

```python
#Print the elements on each index
print(f'The same element using negative and positive indexing: \n Positive: {L[0]}\n Negative: {L[-3]}')
print(f'The same element using negative and positive indexing: \n Positive: {L[1]}\n Negative: {L[-2]} ')
print(f'The same element using negative and postive indexing: \n Positive: {L[2]}\n Negative: {L[-1]}')
```

```python
#Sample List
L = ['Michael Jackson', 10.1, 1982, "MJ",1]
#List slicing
L[3:5]
```

```python
#Using extend to add elements to list
L = ['Michael Jackson',2]
L.extend(['pop',10])
L
```

```python
#Using append to add elements to list
L = ['Michael Jackson', 10.2]
L.append(['pop',10])
L
```

```python
L.append('a')
L
```

```python
#Change the element based on the index
A = ['disco',10,1.2]
print('Before Change:', A)
A[0] = 'hard rock'
print(f'After Change:{A}')
```

```python
#Delete the element based on the index
print(f'Before Change: {A}')
del(A[0])
print('After Change:', A)
```

```python
#Split the string, default is by space
'hard rock'.split()
```

```python
#Split the string by comma
'A,B,C,D'.split(',')
```

```python
#Copy (copy by reference) the list A
A = ['hard rock',10, 1.2]
B = A
print(f'A:{A}')
print(f'B:{B}')
```

```python
#Examine the copy by reference

print(f'B[0]:{B}')
A[0] = 'Banana'
print(f'B[0]:{B}')
```

```python
#Clone  (Clone by Value) the list A
B = A[:]
B
```
